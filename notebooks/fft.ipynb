# FFT Demo
Compute FFT of a simple signal and visualize its magnitude spectrum.

```python
import math
import matplotlib.pyplot as plt
from fft.algorithm import fft

n = 256
x = [math.sin(2*math.pi*5*i/n) + 0.5*math.sin(2*math.pi*12*i/n) for i in range(n)]
X = fft(list(map(complex, x)))
freqs = list(range(n))

plt.figure(figsize=(6,3))
plt.plot(freqs[:n//2], [abs(X[k]) for k in range(n//2)])
plt.title('FFT magnitude (first half)')
plt.xlabel('bin')
plt.ylabel('|X[k]|')
plt.tight_layout()
plt.show()
```

# Gradient Descent Demo
Minimize f(x) = (x-3)^2 + 1 and visualize the parameter updates.

```python
import matplotlib.pyplot as plt
from gradient_descent.algorithm import gradient_descent

f = lambda x: (x-3)**2 + 1
df = lambda x: 2*(x-3)
x0 = 10.0
lr = 0.2

# manual track of iterations for visualization
xs = [x0]
for _ in range(30):
    xs.append(xs[-1] - lr*df(xs[-1]))

plt.figure(figsize=(6,3))
plt.plot(range(len(xs)), xs, marker='o')
plt.title('Gradient descent iterations')
plt.xlabel('step')
plt.ylabel('x')
plt.tight_layout()
plt.show()
```
